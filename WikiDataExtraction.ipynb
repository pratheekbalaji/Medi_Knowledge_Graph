```python
###wikidata disease data

import pandas as pd
import json

wikidata=pd.read_csv('/Users/sharadsharma/Documents/KG/Project/WikiDataDiseaseData.csv')

wikidata=wikidata.rename(columns={'disease_uriLabel':'Disease','HealthSpecialtyLabel':'HealthSpecialty',
                           'TreatmentDrugsLabel':'TreatmentDrugs','CauseLabel':'Cause','SymptonsLabel':'Symptons'})        
        

data_dic={}
wikidata=wikidata.fillna(-1)

for i,row in wikidata.iterrows():
    
    if row['disease_uri'] not in data_dic:
        data_dic[row['disease_uri']]={}
   
    if 'DiseaseURI' not in data_dic[row['disease_uri']]:
        data_dic[row['disease_uri']]['DiseaseURI']=row['disease_uri']
    
    if 'Disease' not in data_dic[row['disease_uri']]:
        data_dic[row['disease_uri']]['Disease']=row['Disease']
    
    if 'HealthSpecialty' not in data_dic[row['disease_uri']]:
        data_dic[row['disease_uri']]['HealthSpecialty']=[row['HealthSpecialty']]
    else:
        data_dic[row['disease_uri']]['HealthSpecialty'].append(row['HealthSpecialty'])
    
    if 'TreatmentDrugs' not in data_dic[row['disease_uri']]:
        #print('yes')
        data_dic[row['disease_uri']]['TreatmentDrugs']=[row['TreatmentDrugs']]
    else:
        #print('n0')
        data_dic[row['disease_uri']]['TreatmentDrugs'].append(row['TreatmentDrugs'])
        
    if 'Cause' not in data_dic[row['disease_uri']]:
        data_dic[row['disease_uri']]['Cause']=[row['Cause']]
    else:
        data_dic[row['disease_uri']]['Cause'].append(row['Cause'])
        
    if 'Symptons' not in data_dic[row['disease_uri']]:
        data_dic[row['disease_uri']]['Symptons']=[row['Symptons']]
    else:
        data_dic[row['disease_uri']]['Symptons'].append(row['Symptons'])

for x in data_dic:
 
    
    data_dic[x]['HealthSpecialty']=list(set(data_dic[x]['HealthSpecialty']))
    data_dic[x]['TreatmentDrugs']=list(set(data_dic[x]['TreatmentDrugs']))
    data_dic[x]['Cause']=list(set(data_dic[x]['Cause']))
    data_dic[x]['Symptons']=list(set(data_dic[x]['Symptons']))
    
    if -1.0 in data_dic[x]['HealthSpecialty']:
        data_dic[x]['HealthSpecialty'].pop()
    if -1.0 in data_dic[x]['TreatmentDrugs']:
        data_dic[x]['TreatmentDrugs'].pop()
    if -1.0 in data_dic[x]['Cause']:
        data_dic[x]['Cause'].pop()
    if -1.0 in data_dic[x]['Symptons']:
        data_dic[x]['Symptons'].pop()
    
with open('/Users/sharadsharma/Documents/KG/Project/WikiDiseaseData.jl','w') as f:
    for x in data_dic:
        json.dump(data_dic[x],f)
        f.write('\n')
```

```python
###wikidata medicine data
wikimed=pd.read_csv('/Users/sharadsharma/Documents/KG/Project/WikiDataMedicineData.csv')
wikimed=wikimed.rename(columns={'medicine_uriLabel':'Medicine','ConditionsTreatedLabel':'ConditionsTreated'})

med_dic={}
wikimed=wikimed.fillna(-1)

for i,row in wikimed.iterrows():
    
    if row['medicine_uri'] not in med_dic:
        med_dic[row['medicine_uri']]={}
   
    if 'MedicineURI' not in med_dic[row['medicine_uri']]:
        med_dic[row['medicine_uri']]['MedicineURI']=row['medicine_uri']
    
    if 'Medicine' not in med_dic[row['medicine_uri']]:
        med_dic[row['medicine_uri']]['Medicine']=row['Medicine']
        
    if 'ConditionsTreated' not in med_dic[row['medicine_uri']]:
        med_dic[row['medicine_uri']]['ConditionsTreated']=[row['ConditionsTreated']]
    else:
        med_dic[row['medicine_uri']]['ConditionsTreated'].append(row['ConditionsTreated'])

for x in med_dic:
    med_dic[x]['ConditionsTreated']=list(set(med_dic[x]['ConditionsTreated']))
    
    if -1.0 in med_dic[x]['ConditionsTreated']:
        med_dic[x]['ConditionsTreated'].pop()
     
with open('/Users/sharadsharma/Documents/KG/Project/WikiMedicineData.jl','w') as f:
    for x in med_dic:
        json.dump(med_dic[x],f)
        f.write('\n')        


```
